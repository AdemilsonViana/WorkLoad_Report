# Adicionar ou alterar uma coluna
### Opera√ß√£o simples de adicionar colunas x Fun√ß√£o para alterar coluna
O Pandas utiliza o Numpy como base, por isso, em certas situa√ß√µes √© possivel aplicar uma opera√ß√£o vetorial sem a necessidade de aplciar fun√ß√µes.
- **Opera√ß√£o simples de adicionar colunas**: Quanto √© poss√≠vel fazer uma opera√ß√£o vetorial, sendo assim n√£o √© necess√°rio utilizar fun√ß√µes. Isso acontece principalmente em opera√ß√µes simples matem√°ticas ou com datas.
- **Opera√ß√£o com fun√ß√£o para adicionar colunas**: Quando h√° opera√ß√µes mais complexas como condicionais, envolve mais de uma coluna ou quando envolve opera√ß√µes de string. Nesses casos √© necess√°rio utilizar fun√ß√£o ou lambda e o comando Apply.

```python
# Adicionar uma coluna (opera√ß√£o simples)
df['duration_x2'] = df['duration'] * 2

# Alterar uma coluna
df['duration'] = df['duration'] * 2
```

```python
# Adicionar uma coluna (opera√ß√£o com fun√ß√£o)
def col_short_type(row):
    if row['type'] == 'studying_calendar':
        return 'studying'
    elif row['type'] == 'working_calendar':
        return 'working'
    elif row['type'] == 'workout_calendar':
        return 'workout'

df['short_type'] = df.apply(col_short_type, axis=1)
```

```python
# Adicionar uma coluna (opera√ß√£o com lambda)
df['short_type'] = df['type'].apply(
    lambda x: 'studying' if x == 'studying_calendar'
    else 'working' if x == 'working_calendar'
    else 'workout'
)
```

# Opera√ß√µes de texto
(substituir texto, separar coluna, juntar colunas, extrair texto)
- str.replace()
- str.split() utilizado para separar um texto
- str.join() utilizado para juntar um texto
- str[] Utilizado como auxiliar para a fun√ß√£o split e joint selecionando a indexa√ß√£o
- essas duas fun√ß√µes s√£o suficientes para fazer diversos tipos de manipula√ß√£oes de texto

```python
# Substituir texto
df['type'] = df['type'].str.replace('_calendar', '', regex=False)
```

```python
# Selecionar um texto do texto, Separar e juntar o texto
df['ANO'] = df['date'].str.split('/').str[2]
df['MES'] = df['date'].str.split('/').str[1]
df['DIA'] = df['date'].str.split('/').str[0]
df['MES-ANO'] = df['date'].str.split('/').str[-2:].str.join('-')
df['ANO-MES'] = df['ANO'] + '-' +df['MES']
```

```python
# uppercase e lower case
df['type'] = df['type'].str.upper()
df['type'] = df['type'].str.lower()
```

```python
# Remover acentos
df['type'] = df['type'].str.replace('calendar', 'calend√°rio', regex=False)

from Functions.data_processing.data_processing import remover_acentos

df['type'] = df['type'].apply(remover_acentos)
```

# Opera√ß√µes de date e Datetime
- em geral as opera√ß√µes utilizadas para as colunas de datetime est√£o em `df['coluna'].dt.`

```python
# ano
df['year'] = df['dt_start'].dt.year
# mes
df['month'] = df['dt_start'].dt.month
# nome mes
df['month_name'] = df['dt_start'].dt.month_name('en')
# dia do m√™s
df['month_day'] = df['dt_start'].dt.day
# dia da semana
df['weekday'] = df['dt_start'].dt.dayofweek
# trimestre
df['quarter'] = df['dt_start'].dt.quarter

# inicio do m√™s 
df['inicio_mes'] = df['dt_start'].dt.to_period('M').dt.start_time()
```

```python
# Retinado fz
df['dt_start'] = df['dt_start'].dt.tz_localize(None)

# Adicionando fz
df['dt_start'] = df['dt_start'].dt.tz_localize('Etc/GMT+3')

# Alterando fz
df['dt_start'] = df['dt_start'].dt.tz_convert('Etc/GMT+3')
```
